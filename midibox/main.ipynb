```python
import librosa
import librosa.display as disp
import matplotlib.pyplot as plt
import numpy as np
```

```python
z,_ = librosa.load('audio/noise.wav')
z = librosa.amplitude_to_db(np.abs(z),ref = np.max)
threshold = min(z)
print(threshold)
```

```python
y, sr = librosa.load('audio/pfft.wav')
y,_ = librosa.effects.trim(y,top_db=35)

plt.plot(y)
plt.show()
```

```python
D = librosa.stft(y)
D = librosa.amplitude_to_db(np.abs(D),ref = np.max)

plt.plot(D)
plt.show()
img = disp.specshow(D)
```

```python
S = librosa.feature.melspectrogram(y=y,sr=sr,n_mels = 128)
S = librosa.amplitude_to_db(np.abs(S),ref = np.max)

plt.plot(S)
plt.show()
img = disp.specshow(S)
```
